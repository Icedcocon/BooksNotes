```c++
//排序：通过投票对团队排名
class Solution {
public:
    string rankTeams(vector<string>& votes) {
        //每个选手对应一个排名列表，第一位记录第一名次数，以此类推
        vector<pair<char,vector<int>>> ranking;
        unordered_map<char,vector<int>> ranking_;
        for(auto player:votes[0])
            ranking_.insert(pair<char,vector<int>>(player,vector<int>(votes[0].size())));
        //统计排名
        for(auto vote:votes)
        {
            int pos=0;
            for(auto player:vote)
                ranking_[player][pos++]++;
        }
        for(auto player:ranking_)
            ranking.push_back(player);
        ranking_.clear();
        //插入排序
        for(int i=1; i<ranking.size(); i++)
        {
            auto key=ranking[i];
            int j;
            for(j=i-1; j>=0 && ranking[j].first>key.first; j--)
                ranking[j+1]=ranking[j];
            ranking[j+1]=key;
        }
        //基数排序(获得pos等奖的次数)
        vector<vector<pair<char,vector<int>>>> buckets(votes.size()+1);//注意这里应该是投票人数量+1而非选手数量
        for(int pos=votes[0].size()-1; pos>=0; pos--)//从最次要位开始排序
        {
            //挂桶
            for(auto player:ranking)
                buckets[player.second[pos]].push_back(player);
            //摘桶
            int count=0;
            for(int i=buckets.size()-1; i>=0; i--)//从大到小排序
            {
                for(auto player:buckets[i])
                    ranking[count++]=player;
                buckets[i].clear();
            }
            
        }
        //输出
        string temp;
        for(auto player:ranking)
            temp.push_back(player.first);
        return temp;
    }
};
```

```c++
//树：重建二叉树
/**
 * Definition for a binary tree node.
 * struct TreeNode {
 *     int val;
 *     TreeNode *left;
 *     TreeNode *right;
 *     TreeNode(int x) : val(x), left(NULL), right(NULL) {}
 * };
 */
class Solution {
public:
    TreeNode* buildTree(vector<int>& preorder, vector<int>& inorder) {
        int preorder_root=0;
        return mybuildTree(preorder, inorder,preorder_root, 0, inorder.size()-1);
    }
TreeNode* mybuildTree(vector<int>& preorder, vector<int>& inorder,int& preorder_root, int low, int high) {
        if(low>high)
            return nullptr;

        // 先把根节点建立出来
        TreeNode* root = new TreeNode(preorder[preorder_root]);

        //在inorder中找到根
        int inorder_root=0;
        while(preorder[preorder_root] != inorder[inorder_root])
            inorder_root++;

        //指向先序遍历序列下一节点，寻找左右孩子
        preorder_root++;

        //判断有无左孩子
        if(inorder_root>low){//有左孩子
            root->left=mybuildTree(preorder,inorder,preorder_root, low, inorder_root-1);
        }

        //判断有无右孩子
        if(inorder_root<high){//有左孩子
            root->right=mybuildTree(preorder,inorder,preorder_root, inorder_root+1, high);
        }
        return root;
    }
};
```
