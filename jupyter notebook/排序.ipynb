```c++
//最大子数组
class Solution {
public:
    int maxSubArray(vector<int>& nums) {
        return FindSubarray(nums, 0, nums.size()-1);
    }
    int FindSubarray(const vector<int>& nums, int low, int high)
    {
        if(low==high)
            return nums[low];
        else
        {
            int mid = (low+high)/2;
            return max(FindCrossingSubarray(nums, low, high, mid),max(FindSubarray(nums, low, mid),FindSubarray(nums, mid+1, high)));
        }
    }
    int FindCrossingSubarray(const vector<int>& nums, int low, int high, int mid)
    {
        int lmax=INT_MIN, rmax=INT_MIN;
        int sum=0;
        for(int i=mid; i>=low; i--)
            if(lmax<(sum+=nums[i])) lmax=sum;
        sum = 0;
        for(int i=mid+1; i<=high; i++)
            if(rmax<(sum+=nums[i])) rmax=sum;
        return lmax+rmax;
    }
};
```

```c++
//折半插入排序
class Solution {
public:
    void InsertSort(vector<int>& nums) {
        int i, j, low, high, mid, key;
        for(i=1; i<nums.size(); i++){ //依次将A[1]～A[n]插入前面已排序元素
            key = nums[i];            //将A[i]暂存到key
            low=0; high=i-1;          //high必须是i-1，不能是i
            while(low<=high){
                mid = (low+high)/2;
                if(nums[mid]>key) high = mid-1;
                else low = mid+1;     //当L[mid]==key时，mid位于左侧不移动的部分
            }
            for(j=i-1; j>=low; j--)
                nums[j+1]=nums[j];    //统一后移一位，空出插入位置
            nums[j+1]=key;
        }
    }
};
```

```c++
//希尔排序
class Solution
{
    public:
    void ShellSort(vector<int>& nums) 
    {
        for(int dk=nums.size()/2; dk>0; dk=dk/2)
        {
            for(int i=dk; i<nums.size(); i++)
            {
                int j, key=nums[i];
                for(j=i-dk; j>=0 && nums[j]>key; j-=dk)
                    nums[j+dk]=nums[j];
                nums[j+dk]=key;
            }
        }
    }
};
```

```c++
//冒泡排序
class Solution
{
    public:
    void BubbleSort(vector<int>& nums)
    {
        bool flag;
        for(int n=nums.size(); n>1; n--)
        {
            flag=false;
            for(int i =1; i<n; i++)
                if(nums[i]<nums[i-1])
                {
                    swap(nums[i], nums[i-1]);
                    flag=true;
                }
            if(flag==false) return;
        }
    }
};
```

```c++
//快速排序《数据结构》版
#include<iostream>
#include<vector>

using namespace std;

class Solution {
public:
    void QuickSort(vector<int> &nums, int low, int high){
        if(low<high){//递归跳出条件，对于基础情况不需要操作的分治法必须设置跳出条件；
            int pivotpos = Partition(nums, low, high);
            QuickSort(nums, low, pivotpos-1);
            QuickSort(nums, pivotpos+1, high);
        }
    }
    int Partition(vector<int> &nums, int low, int high){
        int pivot = nums[low];
        while(low<high){
            while(low<high && pivot<=nums[high]) high--;
            nums[low]=nums[high];
            while(low<high && nums[low]<=pivot) low++;
            nums[high]=nums[low];
        }
        nums[low] = pivot;
        return low;
    }
};
```

```c++
//快速排序《算法导论》版
#include<iostream>
#include<vector>

using namespace std;

class Solution {
public:
    void QuickSort(vector<int> &nums, int low, int high){
        if(low<high){//递归跳出条件，对于基础情况不需要操作的分治法必须设置跳出条件；
            int pivotpos = Partition(nums, low, high);
            QuickSort(nums, low, pivotpos-1);
            QuickSort(nums, pivotpos+1, high);
        }
    }
    int Partition(vector<int> &nums, int low, int high){
        int pivot = nums[high];
        int i=low-1, j;
        for(j=low; j<high; j++){
            if(nums[j]<pivot){
                i++;
                swap(nums[i], nums[j]);
            }
        }
        swap(nums[i+1], nums[high]);
        return i+1;
    }
};
```

```c++
//简单选择排序
class Solution {
public:
    void SelectSort(vector<int> &nums)
    {
        for(int i=0; i<nums.size(); i++)
        {
            int min = i;
            for(int j=i+1; j<nums.size(); j++)
            {
                if(nums[j]<nums[min]) min=j;
            }
            if(min!=i) swap(nums[min],nums[i]);
        }
    }
};
```

```c++
//堆排序
class Solution {
public:
    void BuildMaxHeap(vector<int> &nums){
        for(int i = nums.size()/2-1; i>=0; i--){//i为所有需要下坠的非终端结点,从后向前初始值为n/2-1，0也必须包括，因为0也是
            HeadAdjust(nums, i, nums.size());
        }
        return;
    }
    void HeadAdjust(vector<int> &nums, int k, int n){//数组nums，以待调整元素为根的子树的根节点下标k，堆元素个数heapsize
        for(int i=2*k+1; i<n; i=2*i+1){//下标i指向最大的孩子结点
            if(i+1<n && nums[i]<nums[i+1])
                i++;
            if(nums[k]>=nums[i]) break;
            else{
                swap(nums[i],nums[k]);
                k=i;
            }
        }
    }
    void HeapSort(vector<int> &nums){
        BuildMaxHeap(nums);
        for(int i=nums.size(); i>1; i--){//这里只需要排n-1次，堆中只剩1个元素时自然有序
            swap(nums[0],nums[i-1]);
            HeadAdjust(nums, 0, i-1);
        }
    }
};
```

```c++
//优先队列
#include<iostream>
#include<vector>

using namespace std;


class Solution {
public:
    void BuildMaxHeap(vector<int> &nums){
        for(int i = nums.size()/2-1; i>=0; i--){//i为所有需要下坠的非终端结点,从后向前初始值为n/2-1，0也必须包括，因为0也是
            HeadAdjust(nums, i, nums.size());
        }
        return;
    }
    void HeadAdjust(vector<int> &nums, int k, int n){//数组nums，以待调整元素为根的子树的根节点下标k，堆元素个数heapsize
        for(int i=2*k+1; i<n; i=2*i+1){//下标i指向最大的孩子结点
            if(i+1<n && nums[i]<nums[i+1])
                i++;
            if(nums[k]>=nums[i]) break;
            else{
                swap(nums[i],nums[k]);
                k=i;
            }
        }
    }
    void Insert(vector<int> &nums, int x){
        nums.push_back(x);
        int k = nums.size()-1;//记录插入元素位置
        if(nums.size()==1) return;
        for(int p=(k+1)/2-1; p>=0; p=(p+1)/2-1){//p指向插入元素的父结点，不断上升
            if(nums[k]<=nums[p]) break;
            else{
                swap(nums[p],nums[k]);
                k=p;
            }
        }
    }
    int Maximum(vector<int> &nums){
        return nums[0];
    }
    int ExtractMaximum(vector<int> &nums){
        if(nums.size()==0) return false;
        int key =  nums[0];
        swap(nums[0],nums.back());
        nums.pop_back();
        HeadAdjust(nums, 0, nums.size());
        return key;
    }
    void IncreaseKey(vector<int> &nums, int k, int x){
        if(k>nums.size()-1) return false;//k为待调整结点序号
        if(x<nums[k]) return false;//x为调整后的大小
        nums[k]=x;
        for(int p=(k+1)/2-1; p>=0; p=(p+1)/2-1){//p指向插入元素的父结点，不断上升
            if(nums[k]<=nums[p]) break;
            else{
                swap(nums[p],nums[k]);
                k=p;
            }
        }
    }
};
```

```c++
//归并排序
class Solution {
public:
    void MergeSort(vector<int>& nums)
    {
        doMergeSort(nums,0,nums.size()-1);
    }
    void doMergeSort(vector<int>& nums, int low, int high)
    {
        if(low==high) return;
        int mid=(low+high)/2;
        doMergeSort(nums, low,mid);
        doMergeSort(nums, mid+1,high);
        Merge(nums, low, high);
    }
    void Merge(vector<int>& nums, int low, int high)
    {
        int mid=(low+high)/2;
        vector<int> temp(nums);
        int i,j,k;
        for(i=low, j=mid+1, k=i; i<=mid && j<=high; k++)
        {
            if(temp[i]<=temp[j])
                nums[k]=temp[i++];
            else
                nums[k]=temp[j++];
        }
        while(i<=mid)
            nums[k++]=temp[i++];
        while(j<=high)
            nums[k++]=temp[j++];
    }
};
```

```c++
//基数排序
class Solution 
{
    public:
    void RadixSort(vector<int> &nums)
    {
        //BUCKETS称为基数，即每个元组的取值范围，如十进制为0~9
        const int BUCKETS = 10;
        vector<vector<int>> buckets(BUCKETS);//基数为10，因此建立10个桶存放元素
        
        //每个关键字由pos个元组组成，如三位数由个十百3个元组构成
        for(int pos=0; pos<=2; pos++)
        {
            int denomiator=static_cast<int>(pow(10,pos));//分母
            for(auto num:nums)//将每个元素放入对应的桶中
                buckets[(num/denomiator)%10].push_back(num);//根据对应元组的大小排列
            
            int i=0;
            for(auto& bucket:buckets)//这里的范围for循环中auto后要加引用，因为结束后要将桶清空
            {
                for(auto num:bucket)
                    nums[i++] = num;
                bucket.clear();
            }
        }
    }
};
```

```c++
//基数排序（逆序）
class Solution 
{
    public:
    void RadixSort(vector<int> &nums)
    {
        const int BUCKETS=10;
        vector<vector<int>> buckets(BUCKETS);
        
        for(int pos=0; pos<=2; pos++)
        {
            int denomiator=static_cast<int>(pow(10,pos));
            for(auto num:nums)
                buckets[(num/denomiator)%10].push_back(num);
            
            int count=0;
            //摘桶的顺序决定排序方向，从小到大摘桶为升序，从大到小摘桶为降序
            for(int i=buckets.size()-1; i>=0; i--)
            {
                for(auto num:buckets[i])
                    nums[count++]=num;
                buckets[i].clear();
            }
        }
    }
};
```

```c++
//计数排序
class Solution {
public:
    void CountingSort(vector<int> &nums){
        vector<int>C(10,0);//C存放各数字的个数
        vector<int>B(nums.size(),0);//存放调整后的数字
        for(auto num:nums){
            C[num]++;
        }
        for(int i=1; i<C.size(); i++){
            C[i]=C[i]+C[i-1];
        }
        for(int i=nums.size()-1; i>=0; i--){
            B[C[nums[i]]-1]=nums[i];//从0开始的话要在C[i]-1
            C[nums[i]]--;
        }
        for(auto num:B) cout<<num<<" ";
        cout<<endl;
    }
};
```
